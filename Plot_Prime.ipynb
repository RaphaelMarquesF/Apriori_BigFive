# Plotting Primes

Idea: to create an algorithm to generate and plot prime numbers, but the prime numbers will be divided by 2 until is less than 1 and see what happen.

```python
# importing libraries
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
from matplotlib.pyplot import figure
```

```python
# function to verify if a number is prime
# from wikipedia
def is_prime(n: int) -> bool:
    """Primality test using 6k+-1 optimization."""
    if n <= 3:
        return n > 1
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i ** 2 <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
```

```python
#testing
print(is_prime(457))
print(is_prime(5))
```

```python
# creating a list of prime numbers to select the primes
r = 5000
list_A = []
for x in range(r):
  if is_prime(x) == True:
    list_A.append(x)


print(list_A)
```

```python
# function to divide each prime number until reaches a value between 0 and 1
def Divisor(list=[8],y=2):
  list_One = []
  for x in list:
    while x > 1:
      x = x/y
    if x <= 1:
      list_One.append(x) 
    
  return(list_One)
```

```python
# creating a list of these numbers
A= Divisor(list_A)
```

```python
# function to pass this values to a list
def ListGenerator(valor=10):
  generated = []
  for x in range(valor):
    generated.append(x)
  return(generated)
```

```python
# plotting
x = ListGenerator(len(A))
y = A 

figure(figsize=(18, 9), dpi=100)
plt.plot(x, y, 'o', c='black');
```
